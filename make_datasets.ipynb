```python
with open(file='/content/drive/MyDrive/input2_with_label_0714.pickle', mode='rb') as f:
    df = pickle.load(f).reset_index(drop=True)
```

```python
with open(file='/content/drive/MyDrive/5_HTR/3D_input_5HTR_21dim_min_max_64.pickle', mode='rb') as f:
    input_3D_dict = pickle.load(f)
```

```python
X = df
y = df['pChEMBL_value_Ki']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
from tqdm import tqdm
```

```python
train_data={}

for i in tqdm(range(len(X_train))):
    index = X_train.index[i]
    array = input_3D_dict[(df['target_accession'][i], df['chembl_id'][i])]

    train_data[index] = array
```

```python
test_data={}

for i in tqdm(range(len(X_test))):
    index = X_test.index[i]
    array = input_3D_dict[(df['target_accession'][i], df['chembl_id'][i])]

    test_data[index] = array
```

```python
validation_data={}

for i in tqdm(range(len(X_val))):
    index = X_val.index[i]
    array = input_3D_dict[(df['target_accession'][i], df['chembl_id'][i])]

    validation_data[index] = array
```

```python
with open(file='/content/drive/MyDrive/dataset_5_HTR/train/train_data_21_minmax_64.pickle', mode='wb') as f:
    pickle.dump(X_train, f)
```

```python
with open(file='/content/drive/MyDrive/dataset_5_HTR/train/train_3D_data_21_minmax_64.pickle', mode='wb') as f:
    pickle.dump(train_data, f)
```

```python
with open(file='/content/drive/MyDrive/dataset_5_HTR/test/test_3D_data_21_minmax_64.pickle', mode='wb') as f:
    pickle.dump(test_data, f)
```

```python
with open(file='/content/drive/MyDrive/dataset_5_HTR/test/test_data_21_minmax_64.pickle', mode='wb') as f:
    pickle.dump(X_test, f)
```

```python
with open(file='/content/drive/MyDrive/dataset_5_HTR/validation/val_3D_data_21_minmax_64.pickle', mode='wb') as f:
    pickle.dump(validation_data, f)
```

```python
with open(file='/content/drive/MyDrive/dataset_5_HTR/validation/val_data_21_minmax_64.pickle', mode='wb') as f:
    pickle.dump(X_val, f)
```
